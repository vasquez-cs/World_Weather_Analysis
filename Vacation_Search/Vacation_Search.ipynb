## Deliverable 2. Create a Customer Travel Destinations Map.

```python
# Dependencies and Setup
import pandas as pd
import requests
import gmaps
import matplotlib.pyplot as plt
import numpy as np
import os

# Import API key
from config import g_key

# Configure gmaps API key
gmaps.configure(api_key=g_key)
```

```python
# 1. Import the WeatherPy_database.csv file. 
city_data_df = pd.read_csv("C://Users/Carlos/BootCamp/World_Weather_Analysis/Weather_Database/WeatherPy_Database.csv")
city_data_df.head()
```

```python
# 2. Prompt the user to enter minimum and maximum temperature criteria

# Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
```

```python
# 3. Filter the city_data_df DataFrame using the input statements to create a new DataFrame using the loc method.

# Filter the dataset to find the cities that fit the criteria.
preferred_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & \
                                       (city_data_df["Max Temp"] >= min_temp)]
preferred_cities_df.head(10)
```

```python
# 4a. Determine if there are any empty rows.

# used count of DF to see if there were nulls, all numbers are the same, so no nulls present.
preferred_cities_df.count()
```

```python
# 4b. Drop any empty rows and create a new DataFrame that doesn’t have empty rows.
# No nulls above, double checked below with a "isnull.sum" to return the number of missing values in the data set. still
# zero. Printed it out so that i can view it and clean_df at same time.
print(preferred_cities_df.isnull().sum())
# Step 5a provided code refers to "clean_df" so I set it to preferred_cities_df since it was already clean.
clean_df = preferred_cities_df
# printed out the first 10 rows of clean_df
print (clean_df.head(10))
```

```python
# 5a. Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.

#I had to change the "Current Description" value to "Description" so it matched what I had in my DF.
hotel_df = clean_df[["City", "Country", "Max Temp", "Description", "Lat", "Lng"]].copy()

#I then renamed the column to the expected value: "Current Description".
hotel_df.rename(
    columns={"Description":"Current Description"}
          ,inplace=True)


# 5b. Create a new column "Hotel Name"
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# 6a. Set parameters to search for hotels with 5000 meters.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

# 6b. Iterate through the hotel DataFrame.

# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    
    try:    
        # Get the latitude and longitude.
        lat = row["Lat"]
        lng = row["Lng"]

        # 6c. Get latitude and longitude from DataFrame

        # Add the latitude and longitude to location key for the params dictionary.
        params["location"] = f"{lat},{lng}"    

        # 6d. Set up the base URL for the Google Directions API to get JSON data.

        # Use the search term: "lodging" and our latitude and longitude.
        base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

        # 6e. Make request and retrieve the JSON data from the search.

        # Make request and get the JSON data from the search.
        hotels = requests.get(base_url, params=params).json()

        # 6f. Get the first hotel from the results and store the name, if a hotel isn't found skip the city.

        # Grab the first hotel from the results and store the name.
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
# 7. Drop the rows where there is no Hotel Name.

#check for missing records.
print(hotel_df.head(10))

# I couldnt get rid of nulls at first so referenced the following site to see how to replace my "Hotel Name" blanks 
# with nan which i could remove with a dropna function: 
# https://stackoverflow.com/questions/55426692/how-to-handle-empty-cells-that-pandas-does-not-recognize
hotel_df['Hotel Name']=hotel_df['Hotel Name'].str.strip().replace('',np.nan)

#confirm NaN are now present under "Hotel Name" 
hotel_df.head(10)

# Use dropna function to ensure all NaN are dropped from data frame, set it to new df called hotel_df_clean
hotel_df_clean = hotel_df.dropna()

# confirm that the drop NaN worked.
print(hotel_df_clean.head(10))
```

```python
# 8a. Create the output File (CSV)
output_data_file = "C://Users/Carlos/BootCamp/World_Weather_Analysis/Vacation_Search/WeatherPy_vacation.csv"
# 8b. Export the City_Data into a csv
hotel_df_clean.to_csv(output_data_file, index_label="City_ID")
```

```python
# 9. Using the template add city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description}</dd> and <dd>{Max Temp} °F</dd>
</dl>
"""

# 10a. Get the data from each row and add it to the formatting template and store the data in a list.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df_clean.iterrows()]

# 10b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = hotel_df_clean[["Lat", "Lng"]]
```

```python
# 11a. Add a marker layer for each city to the map. 

# Add a heatmap of temperature for the vacation spots and a pop-up marker for each city.
locations = hotel_df_clean[["Lat", "Lng"]]
max_temp = hotel_df_clean["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)

# 11b. Display the figure
fig
```
